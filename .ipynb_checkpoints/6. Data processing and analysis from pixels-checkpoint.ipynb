### Read raster pixels and extract data

```python
with rio.open(cor_path+'ATSE20201000__LE07_L1TP_192027_20100429_20161214_01_T1_B1_sre.tif') as src:

    data = src.read(1)
    print(data[data>0])
    show(src)
```

```python
pixel_files = os.listdir()

li = []

for file in track(pixel_files):
    
    try:
    
        with rio.open(file) as src:

            data = src.read(1)
            size = data.size
            data = data[data>0][0:9]

            sid = src.name.split('__')[0]

            if 'T1' in file:
                band = src.name.split('T1_')[1].replace('.tif','')
                eid = src.name.split('__')[1][0:40]
            else:
                band = src.name.split('T2_')[1].replace('.tif','')
                eid = src.name.split('__')[1][0:40]


            df1 = pd.DataFrame(data.flatten()).transpose()
            df2 = pd.DataFrame([sid,eid,band,size]).transpose().rename(columns={0:'sid',1:'eid',2:'band',3:'size'})
            df = pd.concat([df2,df1],axis=1)


            li.append(df)
        
    except:
        print(f'Cannot read tif: {file}')


final_df2 = pd.concat(li, axis=0, ignore_index=True)

final_df2.head(2)
```

#### Extract pixel sizes

```python
df_sizes = final_df2['size'].reset_index().drop_duplicates()
df_sizes.head(2)
df_sizes.to_csv("C:/Users/User/University College London/O'Sullivan, Aidan - SDG6extracted_pixels_count.csv")
```

```python
final_df = final_df2.drop('size',axis=1).set_index(['sid','eid','band'])
final_df.columns = ['v0','v1','v2','v3','v4','v5','v6','v7','v8']

df = final_df.stack().reset_index().rename(columns={0:'value'})

## Create ac field
df['ac'] = np.nan
df.loc[df.band.str.contains('sre'),'ac'] = 'sre'
df.loc[df.band.str.contains('tre'),'ac'] = 'tre'
df['ac'] = df.ac.interpolate(method='ffill')

## remove ac from band field
df['band'] = [ i.replace('_sre','').replace('_tre','').replace('_bt','') for i in df.band ]

## unstack to wide bands
df = df.set_index(['sid','eid','band','level_3','ac']).unstack(2)
df = df['value']
df = df[['B1','B2','B3','B4','B5','cirrus','cloud']]
df.head()
```

### Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
from sklearn.utils import check_array
```

```python
def plot_feature_imp():
    plt.title('Feature Importances')
    plt.barh(range(len(fi)), fi['importance'], color='b', align='center')
    plt.yticks(range(len(fi)), fi['feature'])
    plt.xlabel('Relative Importance')
    
    return plt.figure()
```

```python
def fit_and_evaluate_reg(model):
    
    # setup k-fold
    kf = KFold(n_splits=2)
    
    # add df for performance measures
    model_p = pd.DataFrame()
    
    for fold, (train_index, test_index) in enumerate(kf.split(X,y), 1):
        X_train = X.iloc[train_index,:]
        y_train = y[train_index]
        X_test = X.iloc[test_index,:]
        y_test = y[test_index]
    
        # scale data
        sc = StandardScaler()
        X_train = sc.fit_transform(X_train)
        X_test = sc.transform(X_test)
    
        # Train the model
        mod_fit = model.fit(X_train, y_train)
    
        # Predict y on 
        y_pred = model.predict(X_test)
    
        #evaluate model
        
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        model_p = pd.DataFrame({"MAE": [mae], "MSE": [mse],"R_squared":[r2]})
        
        # plot distributions of predictions
        sn.scatterplot(y_pred,y_test)
    
        # Add feature importance table for random forest

        if model==rf:
            fi = pd.DataFrame(data=np.column_stack([X.columns,model.feature_importances_]),
                  index=None,columns=['feature','importance']).sort_values('importance')
        else: 
            fi = []
    
    # Return the performance metric    
    return model_p, mod_fit,fi
```

```python
warnings.filterwarnings('ignore')

## set X and Y

x = sdata.dropna().set_index(['sid','dt'])

X = x.loc[:,['gob']]
y = x['value']

### run selected models

lr = LinearRegression()
lr_per,mod_fit,fi = fit_and_evaluate_reg(lr)

print('\n','Linear Regression Performance: GOB as predictor')
display(lr_per)
```
